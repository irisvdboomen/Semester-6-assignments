We start by creating an empty list. Then, we iterate over the words_no_punc list and check if the word is not in the stopwords list. If the word is not in the stopwords list, we append it to the filtered_words list, all capital letters are lowercased now (.lower()). Finally, we print the filtered_words list and its length.

```python
words_no_punc = []

for w in words:
    if w.isalpha(): 
        words_no_punc.append(w.lower()) 

print(words_no_punc)
print("\n")
print(len(words_no_punc)) 
```

```python
from nltk.corpus import stopwords 

stopwords = stopwords.words("english")
print(stopwords)
```

Here the stopwords are removed from the text.

```python
clean_words = [] 

for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)

print(clean_words) 
print("\n")
print(len(clean_words))
```

The following plot show the most common words, excluding the stopwords.

```python
fdist = FreqDist(clean_words)
fdist.plot(10)
```

### Stemming

Now we will use the Porter Stemmer algorithm to stem the words in the word list. With stemming, we can reduce the words to their root form. this is useful for text processing and analysis.

```python
from nltk.stem import PorterStemmer

porter = PorterStemmer()

# word_list = ["Study", "Studying", "Studies", "Studied"] # word list - example 1
word_list = ["studies", "leaves", "decreases", "plays"] # word list - example 2

for w in word_list:
    print(porter.stem(w))
```

The SnowballStemmer goes further in cutting down words to their roots than the PorterStemmer. It's newer and cuts off more from word endings, making it a stronger choice for breaking down words.

```python
from nltk.stem import SnowballStemmer 

snowball = SnowballStemmer("english") 

word_list = ["Study", "Studying", "Studies", "Studied"] # word list - example 1

for w in word_list:
    print(snowball.stem(w)) 
```

Now we will use the WordNetLemmatizer class to lemmatize words. lemmatization is the process of converting a word to its base form. e.g. studies -> study, leaves -> leaf, decreases -> decrease, plays -> play

```python
from nltk.stem import WordNetLemmatizer 

# nltk.download("wordnet")

lemmatizer = WordNetLemmatizer() 
print(lemmatizer.lemmatize("studies")) 
```

When we set pos="v" for lemmatization, we're telling it to treat words as verbs. This means it will only change words if they can be turned into a simpler verb form. If a word is already in its simplest verb form, it won't change.

```python
lemma = WordNetLemmatizer()
word_list = ["Study", "Studying","Studies", "Studied"] 

for w in word_list:
    print(lemma.lemmatize(w, pos="v")) # lemmatize the words, poss = "v" means verbs
```

### Chunking <br>
Chunking means to extract meaningful phrases from unstructured text. It works on top of PoS tagging. It takes PoS tags as input and provides chunks as outputs. It literally means "a group of words, which breaks simple text into phrases that are more meaningful than individual words." <br>
There are five significant categories of phrases:<br>
<ul>
    <li>Noun Phrases (NP)</li>
    <li>Verb Phrases (VP)</li>
    <li>Adjective Phrases (ADJP)</li>
    <li>Adverb Phrases (ADVP)</li>
    <li>Prepositional Phrases (PP)</li>
</ul><br>
Phrase structure rules:
<ul>
    <li>S (sentence) -> NP VP</li>
    <li>NP -> {Determiner, Noun, Pronoun, Proper name}</li>
    <li>VP -> V (NP)(PP)(Adverb)</li>
    <li>PP -> Pronoun (NP)</li>
    <li>AP -> Adjective (PP)</li>
</ul>

```python
grammar = "NP: {<DT>?<JJ>*<NN>}" # grammar
parser = nltk.RegexpParser(grammar) # create a parser
output = parser.parse(tagged_words) # parse the tagged words
print(output) # print the output

output.draw() # draw the output
```

### Chinking <br>
In complex extractions, it is possible that chunking can output unuseful data. We can then use chinking to exclude some parts from that chunked text.

```python
grammar = r""" NP: {<.*>+}
                }<JJ>+{""" 

parser = nltk.RegexpParser(grammar) # create a parser
output = parser.parse(tagged_words) # parse the tagged words
print(output) # print the output
output.draw() # draw the output
```

### Named Entity Recognition (NER) <br>
NER can automatically scan entire articles and pull out some fundamental entities like people, organizations, places, date, time, money, and GPE discussed in them. A few use-cases are:
<ul>
    <li>Content classification for news channels</li>
    <li>Summarizing resumes</li>
    <li>Optimizing search engine algorithms</li>
    <li>Recommender systems</li>
    <li>Customer supports</li>
</ul>
<br>
When using NER, you can set binary to true or false.
<ul>
    <li>binary = True</li>
    When the binary value is set to true, then it will only show whether a particular entity is a named entity or not. It will not show any further details on it.
    <li>binary = False</li>
    When the binary value is set to false, it shows in detail the type of named entities.
</ul>

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

<b>binary = True</b>

```python
sentence = "Mr. Smith made a deal on a beach of Switzerland near WHO."

tokenized_words = word_tokenize(sentence) # tokenize the sentence

for w in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words) # tag the tokenized words

# print(tagged_words)

N_E_R = nltk.ne_chunk(tagged_words, binary=True) # named entity recognition
print(N_E_R) # print the named entity recognition
N_E_R.draw() # draw the named entity recognition
```

<b>binary = False</b>

```python
sentence = "Mr. Smith made a deal on a beach of Switzerland near WHO."

tokenized_words = word_tokenize(sentence) # tokenize the sentence

for w in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words) # tag the tokenized words

# print(tagged_words)

N_E_R = nltk.ne_chunk(tagged_words, binary=False) # named entity recognition
print(N_E_R) # print the named entity recognition
N_E_R.draw() # draw the named entity recognition
```

### WordNet

A lexical database for the English language. It is part of the NLTK corpus. It is used to find meanings of words, synonyms, antonyms, and more. A few examples are shown below.

```python
# Check how many different definitions of a word exist
from nltk.corpus import wordnet

for words in wordnet.synsets("Fun"):
    print(words)
```

```python
# Check the meaning of those different definitions
for words in wordnet.synsets("Fun"):
    for lemma in words.lemmas():
        print(lemma)
    print("\n")
```

```python
# all details of a word
# for words in wordnet.synsets("Play"): # all of the words that have "Play" in their name

word = wordnet.synsets("Fun")[0]

print(words.name()) # print the name of the words

print(words.definition()) # print the definition of the words

print(words.examples()) # print the examples of the words
```

```python
# all details for all meanings of a word
for words in wordnet.synsets("Fun"):
    print(words.name()) # print the name of the words
    print(words.definition()) # print the definition of the words
    print(words.examples()) # print the examples of the words
    for lemma in words.lemmas():
        print(lemma)
    print("\n")
```

```python
# hypernyms (give us a more abstract term for the word)

# word = wordnet.synsets("Fun")[0]
word = wordnet.synsets("Play")[0]

print(word.hypernyms()) 
```

```python
# hyponyms (give us a more specific term for the word)

word = wordnet.synsets("Play")[0]
word.hyponyms() 
```

```python
# get a name only

word = wordnet.synsets("Play")[0]
print(word.lemmas()[0].name())
```

```python
# synonyms

synonyms = []
for words in wordnet.synsets("Fun"):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())

synonyms
```

```python
# antonyms

antonyms = []

for words in wordnet.synsets("Natural"):
    for lemma in words.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

antonyms
```

```python
# synonyms and antonyms

synonyms = []
antonyms = []

for words in wordnet.synsets("New"):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

print("Synonyms: ")
print(synonyms)
print("\n")
print("Antonyms: ")
print(antonyms)
```

```python
# similarity

word1 = wordnet.synsets("ship", "n")[0]
word2 = wordnet.synsets("boat", "n")[0]
# word3 = wordnet.synsets("car", "n")[0]

print(word1.wup_similarity(word2)) # Wu-Palmer Similarity
```

### Bag of Words

A bag of words is a collection of words that represent a sentence along with the word count where the order of occurences is not relevant. <br>
The structure is:
<ul>
    <li>Raw text</li>
    The original text on which we want to perform analysis.<br><br>
    <li>Clean text</li>
    Removing unnecessary data like punctuation marks and stopwords. The clean text is the text after removing certain words.<br><br>
    <li>Tokenize</li>
    Tokenization represents the sentence as a group of tokens or words.<br><br>
    <li>Building vocab</li>
    Contains total words used in the text after removing unnecessary data.<br><br>
    <li>Generate vocab</li>
    Contains the words along with their frequencies in the sentences.
</ul>

```python
# import required libraries
from sklearn.feature_extraction.text import CountVectorizer

# text for analysis
sentences = ["Jim and Pam travelled by the bus",
             "The train was late",
             "The flight was full. Travelling by flight is expensive"]

# create an object
cv = CountVectorizer()

# generating output for Bag of Words
B_O_W = cv.fit_transform(sentences).toarray()

# total words with their index in model
print(cv.vocabulary_)
print("\n")

# Features
print(cv.get_feature_names_out())
print("\n")

# show the output
print(B_O_W)
```

### TF-IDF

Term Frequency - Inverse Document Frequency. It is a scoring measure generally used in information retrieval (IR) and summarization. The TF-IDF score shows how important or relevant a term is in a given document.

$TF = \frac{\text{{Frequency of the word in the sentence}}}{\text{{Total number of words in the sentence}}}$,    $IDF = \log\left(\frac{\text{{Total number of sentences}}}{\text{{Number of sentences containing that word}}}\right)$

```python
from sklearn.feature_extraction.text import TfidfVectorizer

sentences = ['This is the first document.','This document is the second document.','And this is the third one.','Is this the first document?']

vectorizer = TfidfVectorizer(norm = None)

X = vectorizer.fit_transform(sentences).toarray()

print(vectorizer.vocabulary_)
print("\n")

print(vectorizer.get_feature_names_out())
print("\n")

print(X)
```
